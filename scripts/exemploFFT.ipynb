# Exemplo de CÃ´mputo de FFT via Python

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir os parÃ¢metros do sinal
fs = 1000       # FrequÃªncia de amostragem (Hz)
f1 = 50         # FrequÃªncia da primeira senoide (Hz)
f2 = 120        # FrequÃªncia da segunda senoide (Hz)
dur = 1         # DuraÃ§Ã£o do sinal (s)

# Gerar o sinal
t = np.arange(0, dur, 1/fs)
x = np.sin(2*np.pi*f1*t) + np.sin(2*np.pi*f2*t)

# Computar a FFT usando NumPy
fft_x = np.fft.fft(x)
freqs = np.fft.fftfreq(len(x), 1/fs)

# Plotar o sinal no domÃ­nio do tempo
plt.subplot(2, 1, 1)
plt.plot(t, x)
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.title('Sinal no domÃ­nio do tempo')

# Plotar a FFT do sinal
plt.subplot(2, 1, 2)
plt.plot(freqs, np.abs(fft_x))
plt.xlim([0, fs/2])
plt.xlabel('FrequÃªncia (Hz)')
plt.ylabel('Magnitude')
plt.title('FFT do sinal')

# Mostrar o grÃ¡fico
plt.tight_layout()
plt.show()
```
