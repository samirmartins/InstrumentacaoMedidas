# Exemplo de Cômputo de FFT via Python

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir os parâmetros do sinal
fs = 1000       # Frequência de amostragem (Hz)
f1 = 50         # Frequência da primeira senoide (Hz)
f2 = 120        # Frequência da segunda senoide (Hz)
dur = 1         # Duração do sinal (s)

# Gerar o sinal
t = np.arange(0, dur, 1/fs)
x = np.sin(2*np.pi*f1*t) + np.sin(2*np.pi*f2*t)

# Computar a FFT usando NumPy
fft_x = np.fft.fft(x)
freqs = np.fft.fftfreq(len(x), 1/fs)

# Plotar o sinal no domínio do tempo
plt.subplot(2, 1, 1)
plt.plot(t, x)
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.title('Sinal no domínio do tempo')

# Plotar a FFT do sinal
plt.subplot(2, 1, 2)
plt.plot(freqs, np.abs(fft_x))
plt.xlim([0, fs/2])
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude')
plt.title('FFT do sinal')

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```
