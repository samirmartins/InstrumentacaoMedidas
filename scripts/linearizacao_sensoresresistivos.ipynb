# LINEARIZAÇÃO DE SENSORES RESISTIVOS NÃO LINEARES

```python
import numpy as np
import matplotlib.pyplot as plt

A = 1e-6; # Constante 1 do sensor
B = 5e3;  # Constante 2 do sensor
T = np.linspace(270,350,100); # Variação do mensurando
Rsensor = A*np.exp(B/T)# Relação do sensor com o mensurando T
Rlin = 50 # Valor do resistor que será utilizado para linearização
Rlinearizado = Rsensor*Rlin/(Rsensor+Rlin) # Associação de Resistores em paralelo


plt.plot(T,Rsensor)
plt.plot(T,Rlinearizado)
plt.ylabel('Resistência Equivalente - R($\Omega$)')
plt.xlabel('Temperatura - Kelvin')
plt.grid(True)
plt.legend(["Sensor Não Lnear", "Sensor Linearizado"], loc ="best") 
plt.show()
```
