# Análise de sensibilidade de Instrumentos Medidores

## Instrumento com característica estática não-linear

```python
import sympy as sp
from sympy.plotting import plot 

x = sp.Symbol('x') 
p = plot(0.005*x**2, sp.diff(0.005*x**2,x), show=False, legend = True, xlim=[0,10],
         xlabel = 'Mensurando (x)', ylabel='y = f(x)') 

p[0].line_color = 'b'
p[1].line_color = 'r' 
p[0].label = 'Saída do Instrumento (y)'
p[1].label = 'Sensibilidade = $\\frac{dy}{dx}$'
p.show()
```

## Instrumento com característica estática linear

```python
import sympy as sp
from sympy.plotting import plot 

x = sp.Symbol('x') 
p = plot(5*x+2, sp.diff(5*x+2,x), show=False, legend = True, xlim=[0,10],
         xlabel = 'Mensurando (x)', ylabel='y = f(x)') 

p[0].line_color = 'b'
p[1].line_color = 'r' 
p[0].label = 'Saída do Instrumento (y)'
p[1].label = 'Sensibilidade = $\\frac{dy}{dx}$'
p.show()
```
