# Divisor de tensão e Sensores Resistivos - Escolha de R1

```python
# Escolha de R1 "muito pequeno"
import numpy as np
import matplotlib.pyplot as plt

Rs = np.linspace(0,100,100);
R1 = 10;
Vi = 10;
Vo = Vi*Rs/(Rs+R1)

plt.plot(Rs,Vo)
plt.ylabel('Tensão de Saída - Vo(Volts)')
plt.xlabel('Variação da Resistência Sensor - Rs ($\Omega$)')
plt.grid(True)
#plt.ylim(0, 50)
plt.show()
```
